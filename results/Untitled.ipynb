```python
import pandas as pd
import numpy as np
import re
```

```python
import os
```

```python
array = np.array([])
for file in os.listdir('../'):
    if file.startswith('CIFAR'):
        print(file)
        table = pd.read_csv(f'../{file}')
        if array.shape[0] == 0:
            array = table.values[:, 1:][None].astype(np.float32)
        else:
            array = np.concatenate([array, table.values[:, 1:][None].astype(np.float32)], axis=0)
            
mean_table = table.copy()
std_table = table.copy()

mean_table.iloc[:, 1:] = array.mean(0)
std_table.iloc[:, 1:] = array.std(0)
```

```python
def rename_table(table):
    table.columns = ["number of epoches", '10', '30', '100', '10', '30', '100']
    table.iloc[:, 0] = [
        "VAE", "IWAE,K= 10", "IWAE,K= 50",
                           "L-MCVAE (Fixed),K= 5", "L-MCVAE (Sigmoidal),K= 5", "L-MCVAE (All learnable),K= 5",
                           "L-MCVAE (Fixed),K= 10", "L-MCVAE (Sigmoidal),K= 10", "L-MCVAE (All learnable),K= 10",
                           "A-MCVAE (5 samples, Fixed)),K= 3", "A-MCVAE (5 samples, Sigmoidal),K= 3", "A-MCVAE (5 samples, All learnable),K= 3",
                           "A-MCVAE (5 samples, Fixed)),K= 5", "A-MCVAE (5 samples, Sigmoidal),K= 5", "A-MCVAE (5 samples, All learnable),K= 5"
                          ]
    return table
```

```python
mean_table = rename_table(mean_table.copy())
std_table = rename_table(std_table.copy())
```

```python
mean_table = mean_table.round(2).astype(str)
std_table = std_table.round(2).astype(str)
```

```python
resulting_table = mean_table.copy()

resulting_table.iloc[:, 1:] = mean_table.values[:, 1:] + " $\pm$ " + std_table.values[:, 1:]
```
